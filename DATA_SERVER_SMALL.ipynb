# SENSORS DATA FOR IOT SMART AIR PROJECT

```python
import pandas as pd

df_CO2 = pd.read_csv("co2_data_2019_ny.csv")

df_SO2 = pd.read_csv("so2_data_2019_ny.csv")
df_NO2 = pd.read_csv("no2_data_2019_ny.csv")
df_O3 = pd.read_csv("o3_data_2019_ny.csv")

print(df_O3.columns)

```

```python
#print(df[['Date','Daily Max 8-hour CO Concentration']].head(5))
df_CO2 = df_CO2.rename(columns={"Daily Max 8-hour CO Concentration": "CO2_D8H_C", 'UNITS':'CO2_U','DAILY_AQI_VALUE':'CO2_D_AQI_V' })
df_SO2 = df_SO2.rename(columns={"Daily Max 1-hour SO2 Concentration": "SO2_D1H_C", 'UNITS':'SO2_U','DAILY_AQI_VALUE':'SO2_D_AQI_V'})
df_NO2 = df_NO2.rename(columns={"Daily Max 1-hour NO2 Concentration": "NO2_D1H_C",  'UNITS':'NO2_U','DAILY_AQI_VALUE':'NO2_D_AQI_V'})
df_O3 = df_O3.rename(columns={"Daily Max 8-hour Ozone Concentration": "O3_D8H_C", 'UNITS':'O3_U','DAILY_AQI_VALUE':'O3_D_AQI_V'})
```

```python
#O3_data = df_O3[['Date','O3_D8h_con', 'O3_UNITS', 'O3_DAILY_AQI_VALUE']].head(200).to_json(orient ='index')


CO2_data = df_CO2[['Date','CO2_D8H_C', 'CO2_U', 'CO2_D_AQI_V']].head(200)
SO2_data = df_SO2[['SO2_D1H_C', 'SO2_U', 'SO2_D_AQI_V']].head(200)
NO2_data = df_NO2[['NO2_D1H_C', 'NO2_U', 'NO2_D_AQI_V']].head(200)
O3_data = df_O3[['O3_D8H_C', 'O3_U', 'O3_D_AQI_V']].head(200)
```

```python
#DATA = CO2_data
DATA= result = pd.concat([CO2_data, SO2_data, NO2_data, O3_data], axis=1)
DATA = DATA.sample(frac=1).reset_index(drop=True)
print(DATA)
```

```python

import json
gas_data_json = DATA.to_json(orient ='index')

#print(gas_data_json)
gas_data_json_fn = json.loads(gas_data_json) 
```

```python

#import pandas as pd

import serial
import time


#CO2_data = df[['Date','Daily Max 8-hour CO Concentration']].head(10)


#

try:
    serialPort = serial.Serial(
    port="COM4", baudrate=40000, bytesize=8, timeout=0, stopbits=serial.STOPBITS_ONE
)
except:
    pass




serialString = ""  # Used to hold data coming over UART
while 1:
    for  (k,v) in gas_data_json_fn.items():
        #print(v)
        # Wait until there is data waiting in the serial buffer
        if serialPort.in_waiting > 0:

            # Read data out of the buffer until a carraige return / new line is found
            serialString = serialPort.readline()
            #serialPort.write(b"Hi How are you \r\n")
            out = bytes(json.dumps(v)+"\r\n", 'utf-8')
            print(out)
            serialPort.write(out)
            
            data = json.loads(json.dumps(v)) 
            print(data['Date'])
            # Print the contents of the serial data
            try:
                print(serialString.decode(encoding='UTF-8'))
            except:
                pass
            
            serialPort.flushInput()
            serialPort.flushOutput()
            time.sleep(2)
```
